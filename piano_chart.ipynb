```python
fig, ax = plt.subplots(figsize=(12, 4))

colors = [
    '#D8ACBA',  # a1 (yes - deeper)
    '#b77b8e',  # a2 (no - lighter)
    '#9A536A',  # a3 (deeper)
    '#7A334A',  # a4 (deeper)
    '#44081B',  # a5 (lighter)
    '#A3B1BF',  # d2 (no - lighter)
    '#35526F',  # d1 (yes - deeper)
    '#318375',  # c2 (no - lighter)
    '#09574A',  # c1 (yes - deeper)
    '#FEA88F',  # e2 (no - lighter)
    '#BB5132',  # e1 (yes - deeper)
    '#F8D273',  # b1 (yes - deeper)
    '#946E0D',  # b2 (no/unknown - lighter)
    '#BA9025',  # b3 (deeper)
]

for i, color in enumerate(colors):
    ax.fill_betweenx([i - 0.5, i + 0.5], -1, len(data[0]), color=color, alpha=0.5)

# Connect black circles with lines in the same column first to ensure they are on top
for col_idx in range(len(data[0])):
    black_circle_indices = [row_idx for row_idx, row in enumerate(data) if row[col_idx] == 1]
    for i in range(len(black_circle_indices) - 1):
        ax.plot(
            [col_idx, col_idx],
            [black_circle_indices[i], black_circle_indices[i + 1]],
            color='#6f748b',
            linewidth=1.5,
            zorder=3  # Ensure lines are on top of empty circles
        )

# Plot the data with perfect squares and smaller circles
for row_idx, row in enumerate(data):
    for col_idx, value in enumerate(row):
        # Draw perfect squares
        square = plt.Rectangle((col_idx - 0.5, row_idx - 0.5), 1, 1, facecolor='none', edgecolor='none', zorder=1)
        ax.add_artist(square)
        # Draw smaller circles
        if value == 1:
            # Selected: Draw a black circle
            circle = plt.Circle((col_idx, row_idx), 0.3, color='#6f748b', zorder=4)
            ax.add_artist(circle)
        else:
            # Not selected: Draw an empty circle
            circle = plt.Circle((col_idx, row_idx), 0.3, facecolor='white', edgecolor='none', zorder=2)
            ax.add_artist(circle)

# Set limits and hide tick marks
ax.set_xlim(-0.5, len(data[0]) - 0.5)
ax.set_ylim(-0.5, len(data) - 0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.grid(False)
ax.set_aspect('equal')  # Ensure perfect aspect ratio
ax.set_title("Perfect Circles and Squares with Smaller Circle Radius")

# Show the plot
dpi = 800
plt.savefig('perfect_circles_squares.png', dpi=dpi)
plt.show()
```
