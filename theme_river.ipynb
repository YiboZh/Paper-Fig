```python
colors = {
    'Roadway Type': '#eda388', #// Roadway Type
    'BAC': '#eccc79', #// BAC
    'eeeeWork Zone': '#afba71',# // Work Zone
    'ffffVulnerable User': '#7c9bbf', #// Vulnerable User
    'lllllDriver Behavior': '#de9cad', #// Driver Behavior
    'zzzOthers': '#B3D2B2' # // Others
}
```

```python
df = pd.DataFrame(data, columns=["Date", "Value", "Category"])
# sort the data by 'Value' column
# df = df.sort_values(by='Category', ascending=False)
df = df.sort_values(by='Value', ascending=True)

df["Date"] = pd.to_datetime(df["Date"])  # Convert to datetime
pivot_data = df.pivot(index="Date", columns="Category", values="Value").fillna(0)
pivot_data = pivot_data.reindex(sorted(pivot_data.columns), axis=1)

# Dates, categories, and values
dates = pivot_data.index
categories = pivot_data.columns
values = pivot_data.values

# Calculate stacked values for the theme river effect
stacked_values = np.cumsum(values, axis=1)

# Create a dense range of dates for smooth transitions
dense_dates = np.linspace(0, len(dates) - 1, 300)

category_to_stacked_values = {
    category: stacked_values[:, i] for i, category in enumerate(categories)
}

# Plot the theme river chart
plt.figure(figsize=(14, 7))
# for i, category in enumerate(['BAC', 'Driver Behavior', 'Work Zone', 'Roadway Type', 'Vulnerable User', 'Others']):
for i, category in enumerate(categories):
    if i == 0:
        # For the first category, interpolate between 0 and its value
        # interp = PchipInterpolator(np.arange(len(dates)), stacked_values[:, i])
        interp = PchipInterpolator(np.arange(len(dates)), category_to_stacked_values[category])
        y_smooth = interp(dense_dates)
        plt.fill_between(dense_dates, 0, y_smooth, label=category, color=colors[category])
    else:
        # For subsequent categories, interpolate between previous and current stacked values
        # interp1 = PchipInterpolator(np.arange(len(dates)), stacked_values[:, i - 1])
        interp1 = PchipInterpolator(np.arange(len(dates)), category_to_stacked_values[categories[i - 1]])
        # interp2 = PchipInterpolator(np.arange(len(dates)), stacked_values[:, i])
        interp2 = PchipInterpolator(np.arange(len(dates)), category_to_stacked_values[category])
        y1_smooth = interp1(dense_dates)
        y2_smooth = interp2(dense_dates)
        plt.fill_between(dense_dates, y1_smooth, y2_smooth, label=category, color=colors[category])

# Adjust the x-axis labels to use the actual dates
plt.xticks(
    np.linspace(0, len(dates) - 1, len(dates)),
    [d.strftime('%Y-%m-%d') for d in dates],
    rotation=45
)

# Customize the chart
plt.title("Theme River Chart with Smooth Curves", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Value", fontsize=12)
# plt.legend(title="Category", loc="upper left")
plt.gca().grid(False)  # Remove grid lines

# Show the plot
plt.tight_layout()

plt.savefig('theme_river_chart.png', dpi=800)
```
