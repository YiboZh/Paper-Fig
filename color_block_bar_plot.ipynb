```python
fig, ax = plt.subplots(figsize=(10, 6))
height = 0.5
y_positions = np.arange(len(categories))

for i, pred_value in enumerate(preds):
    ax.barh(
        y_positions[i],
        pred_value * 12,
        height,
        color="gray",
        alpha=0.3,
        zorder=0,
    )

start_points = np.zeros(len(categories))

for i in range(4):
    widths = data_mixed_values[:, i] * 12
    positive_widths = np.clip(widths, 0, None)
    negative_widths = np.clip(widths, None, 0)
    segment_colors = [color_mapping[legend[j][i]] for j in range(len(legend))]

    ax.barh(
        y_positions,
        positive_widths,
        height,
        left=start_points,
        color=segment_colors,
        alpha=0.4,
        label=f"{legend[0][i]}" if i == 0 else None,
    )

    ax.barh(
        y_positions,
        np.abs(negative_widths),
        height,
        left=start_points + positive_widths,
        color=segment_colors,
        alpha=0.7,
    )

    start_points += positive_widths + np.abs(negative_widths)


for i, pred_value in enumerate(preds):
    ax.plot(
        [pred_value * 12, pred_value * 12],
        [y_positions[i] - height / 2, y_positions[i] + height / 2],
        color="black",
        linewidth=2,
    )
    ax.text(
        max(pred_value * 12 + 0.5, start_points[i]),
        y_positions[i],
        f"{pred_value:.3f}",
        fontsize=9,
        ha="left",
        va="center",
        color="black",
    )

for i, category in enumerate(categories):
    non_zero_segments = [j for j, value in enumerate(data_mixed_values[i]) if value != 0]
    for j, segment_idx in enumerate(non_zero_segments):
        ax.add_patch(
            plt.Rectangle(
                (-1 - j * 0.3, i - height / 2),
                0.22,
                height,
                color=color_mapping[legend[i][segment_idx]],
                alpha=0.7,
            )
        )

for i, section in enumerate(sections):
    section_start = i * 7
    ax.axhline(section_start - 0.5, color="gray", linestyle="--", linewidth=1)
    if 'and' in section:
        section = "Male and\nFemale"
    if 'or' in section:
        section = "Unkonwn or\nNot Provided"
    ax.text(-4, section_start + 3, section, fontsize=10, ha="left", va="center", fontweight="bold", color="black")

for i, category in enumerate(categories):
    ax.text(-1, i, category, fontsize=10, ha="right", va="center", color="black")

ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(True)
ax.spines["bottom"].set_visible(False)
ax.xaxis.set_ticks_position("top")
ax.xaxis.set_label_position("top")
ax.set_xticks([])
ax.set_yticks([])
ax.set_yticklabels([])
ax.set_xlim(-4, 20)
ax.set_title("Age-Group-Specific Contribution Analysis of Factors Impacting Risk Score Severity", fontsize=14)

# ax.legend(loc="upper right", title="Legend")
custom_legends = []

for group_name, mapping in grouped_color_mapping.items():
    custom_legends.append(
        plt.Line2D([0], [0], color="none", label=group_name, linewidth=0)
    )
    for label, color in mapping.items():
        custom_legends.append(
            plt.Line2D([0], [0], color=color, lw=4, label=label)
        )

plt.tight_layout()
plt.show()
```
